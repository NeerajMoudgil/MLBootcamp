```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic= sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.distplot(titanic['fare'])
```

```python
sns.distplot(titanic['fare'],kde=False,color='red')
```

```python
sns.distplot(titanic['fare'],kde=False,color='red',bins=30)
```

```python
sns.boxplot(x='class',y='age',data=titanic) #categorical
```

```python
sns.barplot(x='class',y='age',data=titanic) #categorical
```

```python
sns.countplot(x='class',data=titanic) #categorical
```

```python
sns.heatmap(titanic.corr())
```

```python
sns.clustermap(titanic.corr())
```

### matplotlib

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
